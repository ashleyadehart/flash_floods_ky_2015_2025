```python
import requests
import pandas as pd
import time
```

```python
API_KEY = "YOUR API KEY GOES HERE!"
BASE_URL = "http://api.weatherapi.com/v1/history.json"
```

```python
df = pd.read_csv("data/output/county_city_date_counts.csv")
```

```python
df.head()
```

```python
results = []
```

```python
row = df.iloc[0]

city = row["BEGIN_LOCATION"]
date = row["BEGIN_DATE"]

query = f"{city}, KY"

params = {
    "key": API_KEY,
    "q": query,
    "dt": date
}

response = requests.get(BASE_URL, params=params)
data = response.json()

data
```

```python
day = data["forecast"]["forecastday"][0]["day"]

print(day["avgtemp_f"])
print(day["maxtemp_f"])
print(day["mintemp_f"])
print(day["condition"]["text"])
```

```python
for index, row in df.iterrows():
    
    city = row["BEGIN_LOCATION"]
    county = row["COUNTY_NAME"]
    date = row["BEGIN_DATE"]

    query = f"{city}, KY"

    params = {
        "key": API_KEY,
        "q": query,
        "dt": date
    }

    try:
        response = requests.get(BASE_URL, params=params)
        data = response.json()

        day = data["forecast"]["forecastday"][0]["day"]

        results.append({
            "BEGIN_DATE": date,
            "COUNTY_NAME": county,
            "BEGIN_LOCATION": city,
            "avg_temp_f": day["avgtemp_f"],
            "max_temp_f": day["maxtemp_f"],
            "min_temp_f": day["mintemp_f"],
            "precip_in": day["totalprecip_in"],
            "condition": day["condition"]["text"]
        })

        print("Success:", city)

    except:
        print("Failed:", city)

    time.sleep(1)
```

```python
weather_df = pd.DataFrame(results)
weather_df.head()
```

```python
weather_df.to_csv("data/output/kentucky_weather_top_five_dates.csv", index=False)
```
